```python
# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║  SYMPTOM MODEL V2 — SINGLE TRAINING RUN (NO FOLDS)                         ║
# ║  Architecture: DenseNet + BiLSTM + Attention (same as NB1)                  ║
# ║  100 epochs, synthetic expansion, augmentation, MixUp, label smoothing      ║
# ║  Run on Google Colab GPU. Dataset: symbipredict_2022.csv                    ║
# ╚══════════════════════════════════════════════════════════════════════════════╝

import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
import keras, os, pickle, warnings
warnings.filterwarnings('ignore')
from keras.models import Model
from keras.layers import (Input, Dense, Dropout, Conv1D, Flatten, LSTM,
                          Bidirectional, Attention, concatenate, BatchNormalization)
from keras.regularizers import l2
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping, ModelCheckpoint, Callback
from keras.utils import to_categorical
from keras.losses import CategoricalCrossentropy
from keras.metrics import CategoricalAccuracy
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import (classification_report, confusion_matrix, f1_score,
                             recall_score, top_k_accuracy_score, roc_auc_score,
                             roc_curve, auc as sk_auc)
from itertools import combinations

# ═══════ CONFIG ═══════
DATASET_PATH = '/content/symbipredict_2022.csv'
SAVE_DIR     = '/content/saved_models'
FIG_DIR      = '/content/figures'
RES_DIR      = '/content/results'
for d in [SAVE_DIR, FIG_DIR, RES_DIR]:
    os.makedirs(d, exist_ok=True)

EP            = 100
BS            = 32
L2R           = 1e-3
LABEL_SMOOTH  = 0.05
MIXUP_ALPHA   = 0.2
AUG_PROB_MIN  = 0.35
AUG_PROB_MAX  = 0.70
LR_INIT       = 1e-3
LR_MIN        = 1e-6
WARMUP_EP     = 5
SYNTH_MULT    = 4

ACUTE_LABELS = ['Heart Attack','Pneumonia','Tuberculosis','Typhoid','Hepatitis E',
                'Dengue','Malaria','Jaundice','Paralysis (brain hemorrhage)',
                'Alcoholic Hepatitis','Drug Reaction','Gastroenteritis']

# ═══════ 1. LOAD DATA & BUILD KNOWLEDGE BASE ═══════
print("="*60+"\n  Loading data & building disease knowledge base\n"+"="*60)
dataset = pd.read_csv(DATASET_PATH)
SYMPTOMS = dataset.drop('prognosis', axis=1).columns.tolist()
N_SYMS = len(SYMPTOMS)
X_all = dataset.drop('prognosis', axis=1).values.astype(np.float32)
y_raw = dataset['prognosis'].values
le = LabelEncoder()
y_all = le.fit_transform(y_raw)
NC = len(le.classes_)
print(f"  {X_all.shape[0]} samples, {N_SYMS} features, {NC} classes")

DSYM_FREQ, DSYM_CORE, DSYM_POOL, DSYM_RARE = {}, {}, {}, {}
for disease, group in dataset.groupby('prognosis'):
    X_d = group.drop('prognosis', axis=1).values.astype(float)
    freq = X_d.mean(axis=0)
    DSYM_FREQ[disease] = {i: freq[i] for i in range(N_SYMS) if freq[i] > 0}
    DSYM_CORE[disease] = [i for i in range(N_SYMS) if freq[i] >= 0.80]
    DSYM_POOL[disease] = [i for i in range(N_SYMS) if freq[i] >= 0.10]
    DSYM_RARE[disease] = [i for i in range(N_SYMS) if 0 < freq[i] < 0.40]

# Confusable pairs
CONFUSE_MAP = {}
diseases_list = list(DSYM_POOL.keys())
for d1, d2 in combinations(diseases_list, 2):
    s1, s2 = set(DSYM_POOL[d1]), set(DSYM_POOL[d2])
    if not s1 | s2:
        continue
    j = len(s1 & s2) / len(s1 | s2)
    if j < 0.15:
        continue
    only1, only2 = sorted(s1 - s2), sorted(s2 - s1)
    if d1 not in CONFUSE_MAP or j > CONFUSE_MAP[d1][0]:
        CONFUSE_MAP[d1] = (j, d2, only2)
    if d2 not in CONFUSE_MAP or j > CONFUSE_MAP[d2][0]:
        CONFUSE_MAP[d2] = (j, d1, only1)

CONFUSE_ALL = {d: [] for d in diseases_list}
for d1, d2 in combinations(diseases_list, 2):
    s1, s2 = set(DSYM_POOL[d1]), set(DSYM_POOL[d2])
    if not s1 | s2:
        continue
    j = len(s1 & s2) / len(s1 | s2)
    if j < 0.15:
        continue
    CONFUSE_ALL[d1].append((j, d2, sorted(s2 - s1)))
    CONFUSE_ALL[d2].append((j, d1, sorted(s1 - s2)))
for d in CONFUSE_ALL:
    CONFUSE_ALL[d].sort(key=lambda x: -x[0])

acute_idx = [np.where(le.classes_ == l)[0][0] for l in ACUTE_LABELS if l in le.classes_]
print(f"  Confusable pairs: {sum(len(v) for v in CONFUSE_ALL.values()) // 2}")


# ═══════ 2. SYNTHETIC DATA EXPANSION ═══════
print(f"\n{'='*60}\n  Synthetic data expansion ({SYNTH_MULT}x)\n{'='*60}")

def generate_synthetic_samples(X_disease, disease_name, n_generate, rng):
    freq_map = DSYM_FREQ.get(disease_name, {})
    core = DSYM_CORE.get(disease_name, [])
    pool = DSYM_POOL.get(disease_name, [])
    rare = DSYM_RARE.get(disease_name, [])
    confuse_partners = CONFUSE_ALL.get(disease_name, [])
    samples = []
    for _ in range(n_generate):
        base = X_disease[rng.randint(len(X_disease))].copy()
        strategy = rng.randint(0, 8)
        if strategy == 0:
            on = np.where(base == 1)[0]
            if len(on) > 2:
                keep_frac = rng.uniform(0.40, 0.70)
                n_keep = max(2, int(len(on) * keep_frac))
                ranked = sorted(on, key=lambda j: freq_map.get(j, 0), reverse=True)
                keep = set(ranked[:n_keep])
                for j in on:
                    if j not in keep:
                        base[j] = 0
        elif strategy == 1:
            on = np.where(base == 1)[0]
            if len(on) > 3 and len(core) >= 2:
                n_keep = rng.randint(2, min(5, len(core) + 1))
                keep = set(rng.choice(core, size=min(n_keep, len(core)), replace=False))
                for j in on:
                    if j not in keep:
                        base[j] = 0
        elif strategy == 2:
            on = np.where(base == 1)[0]
            non_core = [j for j in on if j not in set(core)]
            if len(non_core) >= 2:
                for j in core:
                    if j in on:
                        base[j] = 0
                if rare and rng.rand() < 0.5:
                    add = [r for r in rare if base[r] == 0]
                    if add:
                        base[rng.choice(add)] = 1
        elif strategy == 3:
            disease_pool_set = set(pool)
            off = [j for j in range(N_SYMS) if base[j] == 0 and j not in disease_pool_set]
            if len(off) >= 1:
                n_add = rng.randint(1, min(9, len(off) + 1))
                base[rng.choice(off, size=n_add, replace=False)] = 1
        elif strategy == 4:
            if confuse_partners:
                _, _, inject_pool = confuse_partners[rng.randint(len(confuse_partners))]
                cands = [j for j in inject_pool if base[j] == 0]
                if len(cands) >= 1:
                    n_add = rng.randint(1, min(6, len(cands) + 1))
                    base[rng.choice(cands, size=n_add, replace=False)] = 1
        elif strategy == 5:
            on = np.where(base == 1)[0]
            if len(on) > 3:
                n_drop = max(1, int(len(on) * rng.uniform(0.2, 0.4)))
                n_drop = min(n_drop, len(on) - 2)
                base[rng.choice(on, size=n_drop, replace=False)] = 0
            off = np.where(base == 0)[0]
            n_add = rng.randint(1, 4)
            if len(off) >= n_add:
                base[rng.choice(off, size=n_add, replace=False)] = 1
        elif strategy == 6:
            base[:] = 0
            if len(pool) >= 3:
                n_pick = rng.randint(3, min(len(pool) + 1, 15))
                base[rng.choice(pool, size=n_pick, replace=False)] = 1
        elif strategy == 7:
            on = np.where(base == 1)[0]
            off = np.where(base == 0)[0]
            if len(on) > 2 and rng.rand() < 0.7:
                base[rng.choice(on)] = 0
            if len(off) > 0:
                base[rng.choice(off)] = 1
        samples.append(base)
    return np.array(samples, dtype=np.float32)

def expand_dataset(X, y_int, mult=SYNTH_MULT, seed=42):
    rng = np.random.RandomState(seed)
    X_new, y_new = [X.copy()], [y_int.copy()]
    for cls_idx in range(NC):
        mask = y_int == cls_idx
        X_cls = X[mask]
        disease = le.classes_[cls_idx]
        n_gen = len(X_cls) * mult
        X_syn = generate_synthetic_samples(X_cls, disease, n_gen, rng)
        X_new.append(X_syn)
        y_new.append(np.full(len(X_syn), cls_idx, dtype=y_int.dtype))
    X_out = np.concatenate(X_new)
    y_out = np.concatenate(y_new)
    perm = rng.permutation(len(X_out))
    return X_out[perm], y_out[perm]


# ═══════ 3. AUGMENTATION ENGINE ═══════
class AugmentationEngine:
    def __init__(self, rng_seed=42):
        self.rng = np.random.RandomState(rng_seed)

    def augment_batch(self, X_batch, y_batch_int, aug_prob):
        X_aug = X_batch.copy()
        do_aug = self.rng.rand(len(X_aug)) < aug_prob
        for i in range(len(X_aug)):
            if not do_aug[i]:
                continue
            disease = le.classes_[y_batch_int[i]]
            n_augs = 2 if self.rng.rand() < 0.25 else 1
            for _ in range(n_augs):
                s = self.rng.randint(0, 7)
                if s == 0:   self._dropout(X_aug, i, disease, 0.30, 0.60)
                elif s == 1: self._noise(X_aug, i, disease)
                elif s == 2: self._atypical(X_aug, i, disease)
                elif s == 3: self._cross_contam(X_aug, i, disease)
                elif s == 4: self._partial_core(X_aug, i, disease)
                elif s == 5: self._dropout(X_aug, i, disease, 0.15, 0.30)
                elif s == 6: self._random_flip(X_aug, i)
        return X_aug

    def _dropout(self, X, i, disease, lo, hi):
        on = np.where(X[i] == 1)[0]
        if len(on) <= 2: return
        n_drop = min(max(1, int(len(on) * self.rng.uniform(lo, hi))), len(on) - 2)
        X[i, self.rng.choice(on, size=n_drop, replace=False)] = 0

    def _noise(self, X, i, disease):
        dp = set(DSYM_POOL.get(disease, []))
        cands = [j for j in range(N_SYMS) if X[i, j] == 0 and j not in dp]
        if len(cands) < 1: return
        n = self.rng.randint(1, min(8, len(cands) + 1))
        X[i, self.rng.choice(cands, size=n, replace=False)] = 1

    def _atypical(self, X, i, disease):
        fm = DSYM_FREQ.get(disease, {})
        on = np.where(X[i] == 1)[0]
        if len(on) <= 2: return
        core = [j for j in on if fm.get(j, 0) >= 0.80]
        non_core = [j for j in on if fm.get(j, 0) < 0.80]
        if len(non_core) >= 2:
            for j in core: X[i, j] = 0
        elif len(core) > 1:
            X[i, self.rng.choice(core, size=max(1, len(core) // 2), replace=False)] = 0

    def _cross_contam(self, X, i, disease):
        if disease not in CONFUSE_MAP:
            self._dropout(X, i, disease, 0.15, 0.30)
            return
        partners = CONFUSE_ALL.get(disease, [])
        if not partners: return
        _, _, pool = partners[self.rng.randint(len(partners))]
        cands = [j for j in pool if X[i, j] == 0]
        if len(cands) < 1: return
        n = self.rng.randint(1, min(7, len(cands) + 1))
        X[i, self.rng.choice(cands, size=n, replace=False)] = 1

    def _partial_core(self, X, i, disease):
        fm = DSYM_FREQ.get(disease, {})
        on = np.where(X[i] == 1)[0]
        if len(on) <= 3: return
        ranked = sorted(on, key=lambda j: fm.get(j, 0), reverse=True)
        n_keep = self.rng.randint(2, 4)
        keep = set(ranked[:n_keep])
        for j in on:
            if j not in keep: X[i, j] = 0

    def _random_flip(self, X, i):
        on = np.where(X[i] == 1)[0]
        off = np.where(X[i] == 0)[0]
        if len(on) > 2 and self.rng.rand() < 0.6:
            X[i, self.rng.choice(on)] = 0
        if len(off) > 0:
            X[i, self.rng.choice(off)] = 1


# ═══════ 4. MIXUP ═══════
def mixup_batch(X, y_ohe, alpha=MIXUP_ALPHA):
    if alpha <= 0: return X, y_ohe
    lam = np.random.beta(alpha, alpha, size=len(X)).astype(np.float32)
    lam = np.maximum(lam, 1 - lam)
    idx = np.random.permutation(len(X))
    return (lam.reshape(-1, 1) * X + (1 - lam.reshape(-1, 1)) * X[idx],
            lam.reshape(-1, 1) * y_ohe + (1 - lam.reshape(-1, 1)) * y_ohe[idx])


# ═══════ 5. CALLBACKS ═══════
class CosineAnnealingWarmup(Callback):
    def __init__(self, lr_max, lr_min, warmup_ep, total_ep):
        super().__init__()
        self.lr_max = lr_max
        self.lr_min = lr_min
        self.warmup_ep = warmup_ep
        self.total_ep = total_ep
        self.lr_history = []

    def on_epoch_begin(self, ep, logs=None):
        if ep < self.warmup_ep:
            lr = self.lr_min + (self.lr_max - self.lr_min) * (ep / self.warmup_ep)
        else:
            p = (ep - self.warmup_ep) / max(1, self.total_ep - self.warmup_ep)
            lr = self.lr_min + 0.5 * (self.lr_max - self.lr_min) * (1 + np.cos(np.pi * p))
        self.model.optimizer.learning_rate.assign(lr)
        self.lr_history.append(lr)

class TestCB(Callback):
    def __init__(self, Xt, yt):
        super().__init__()
        self.Xt = Xt
        self.yt = yt
        self.tl = []
        self.ta = []

    def on_epoch_end(self, ep, logs=None):
        l, a = self.model.evaluate(self.Xt, self.yt, verbose=0)
        self.tl.append(l)
        self.ta.append(a)


# ═══════ 6. TRAINING GENERATOR ═══════
class AugmentedSequence(keras.utils.Sequence):
    def __init__(self, X, y_int, nc, bs, augmenter, scaler,
                 use_mixup=True, shuffle=True, rng_seed=42):
        self.X_raw = X.copy()
        self.y_int = y_int.copy()
        self.nc = nc
        self.bs = bs
        self.augmenter = augmenter
        self.scaler = scaler
        self.use_mixup = use_mixup
        self.shuffle = shuffle
        self.rng = np.random.RandomState(rng_seed)
        self.indices = np.arange(len(X))
        self.epoch = 0
        if shuffle:
            self.rng.shuffle(self.indices)

    def __len__(self):
        return int(np.ceil(len(self.X_raw) / self.bs))

    def __getitem__(self, idx):
        sl = self.indices[idx * self.bs: min((idx + 1) * self.bs, len(self.X_raw))]
        X_b = self.X_raw[sl].copy()
        y_b = self.y_int[sl].copy()
        prog = min(self.epoch / max(1, EP * 0.6), 1.0)
        aug_prob = AUG_PROB_MIN + (AUG_PROB_MAX - AUG_PROB_MIN) * prog
        X_b = self.augmenter.augment_batch(X_b, y_b, aug_prob)
        X_b = self.scaler.transform(X_b)
        y_ohe = to_categorical(y_b, self.nc)
        if self.use_mixup:
            X_b, y_ohe = mixup_batch(X_b, y_ohe)
        return np.expand_dims(X_b, -1), y_ohe

    def on_epoch_end(self):
        self.epoch += 1
        if self.shuffle:
            self.rng.shuffle(self.indices)


# ═══════ 7. MODEL — SAME AS NB1 ═══════
def DenseNetBlock(inp, gr=24, layers=5):
    for _ in range(layers):
        x = Conv1D(gr, 3, padding='same', activation='relu', kernel_regularizer=l2(L2R))(inp)
        x = BatchNormalization()(x)
        x = Dropout(0.5)(x)
        inp = concatenate([inp, x])
    return inp

def LSTMAttnBlock(inp):
    o = Bidirectional(LSTM(200, return_sequences=True, kernel_regularizer=l2(L2R)))(inp)
    return Attention()([o, o])

def build_model(shape, nc):
    i = Input(shape=shape)
    dn = i
    for _ in range(3):
        dn = DenseNetBlock(dn)
    a1 = Attention()([dn, dn])
    lo = a1
    for _ in range(2):
        lo = LSTMAttnBlock(lo)
    f = concatenate([Flatten()(a1), Flatten()(lo)])
    x = Dense(512, activation='relu', kernel_regularizer=l2(L2R))(f)
    x = Dropout(0.6)(x)
    x = Dense(256, activation='relu', kernel_regularizer=l2(L2R))(x)
    x = Dropout(0.6)(x)
    x = Dense(128, activation='relu', kernel_regularizer=l2(L2R))(x)
    x = Dropout(0.6)(x)
    o = Dense(nc, activation='softmax')(x)
    m = Model(i, o)
    m.compile(
        optimizer=Adam(LR_INIT),
        loss=CategoricalCrossentropy(label_smoothing=LABEL_SMOOTH),
        metrics=[CategoricalAccuracy(name='accuracy')]
    )
    return m


# ═══════ 8. TEMPERATURE SCALING ═══════
def find_optimal_temperature(model, X_val, y_val_int, temps=np.arange(0.5, 5.0, 0.1)):
    logits = model.predict(X_val, verbose=0)
    best_t, best_nll = 1.0, float('inf')
    for t in temps:
        scaled = np.exp(np.log(logits + 1e-12) / t)
        scaled = scaled / scaled.sum(axis=1, keepdims=True)
        nll = -np.mean(np.log(scaled[np.arange(len(y_val_int)), y_val_int] + 1e-12))
        if nll < best_nll:
            best_nll = nll
            best_t = t
    return best_t

def apply_temperature(probs, temperature):
    if temperature == 1.0: return probs
    scaled = np.exp(np.log(probs + 1e-12) / temperature)
    return scaled / scaled.sum(axis=1, keepdims=True)


# ═══════ 9. SPLIT DATA ═══════
print(f"\n{'='*60}\n  Splitting data (70% train, 10% val, 20% test)\n{'='*60}")
X_trainval, X_test, y_trainval, y_test = train_test_split(
    X_all, y_all, test_size=0.20, stratify=y_all, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval, test_size=0.125, stratify=y_trainval, random_state=42)
# 0.125 of 80% = 10% of total

print(f"  Train: {len(X_train)}  Val: {len(X_val)}  Test: {len(X_test)}")

y_val_ohe = to_categorical(y_val, NC)
y_test_ohe = to_categorical(y_test, NC)


# ═══════ 10. EXPAND & PREPARE ═══════
X_train_exp, y_train_exp = expand_dataset(X_train, y_train, mult=SYNTH_MULT, seed=42)
print(f"  Expanded training: {len(X_train)} → {len(X_train_exp)} samples")

# Fit scaler on REAL training data only
scaler = StandardScaler()
scaler.fit(X_train)

X_val_scaled = np.expand_dims(scaler.transform(X_val), -1)
X_test_scaled = np.expand_dims(scaler.transform(X_test), -1)

# Build generator
augmenter = AugmentationEngine(rng_seed=42)
train_gen = AugmentedSequence(
    X=X_train_exp, y_int=y_train_exp, nc=NC, bs=BS,
    augmenter=augmenter, scaler=scaler,
    use_mixup=True, shuffle=True, rng_seed=42
)


# ═══════ 11. TRAIN ═══════
print(f"\n{'='*60}\n  TRAINING — 100 epochs, single run\n  Architecture: DenseNet + BiLSTM + Attention (NB1)\n{'='*60}")

model = build_model((N_SYMS, 1), NC)
model.summary()

ckpt_path = f'{SAVE_DIR}/final_symptom_model.keras'
test_cb = TestCB(X_test_scaled, y_test_ohe)
cosine_lr = CosineAnnealingWarmup(LR_INIT, LR_MIN, WARMUP_EP, EP)

history = model.fit(
    train_gen,
    validation_data=(X_val_scaled, y_val_ohe),
    epochs=EP,
    callbacks=[
        EarlyStopping(monitor='val_loss', patience=15,
                      restore_best_weights=True, verbose=1),
        ModelCheckpoint(ckpt_path, monitor='val_accuracy',
                        save_best_only=True, mode='max', verbose=1),
        cosine_lr,
        test_cb,
    ],
    verbose=1,
)


# ═══════ 12. TEMPERATURE SCALING ═══════
print(f"\n{'='*60}\n  Temperature Scaling\n{'='*60}")
opt_temp = find_optimal_temperature(model, X_val_scaled, y_val)
print(f"  Optimal temperature: {opt_temp:.2f}")


# ═══════ 13. EVALUATE ON TEST SET ═══════
print(f"\n{'='*60}\n  Test Set Evaluation\n{'='*60}")
tl, ta = model.evaluate(X_test_scaled, y_test_ohe, verbose=0)
yp_raw = model.predict(X_test_scaled, verbose=0)
yp = apply_temperature(yp_raw, opt_temp)
ypc = np.argmax(yp, 1)

f1 = f1_score(y_test, ypc, average='macro')
t3 = top_k_accuracy_score(y_test, yp, k=3)
t5 = top_k_accuracy_score(y_test, yp, k=5)
try:
    auc_val = roc_auc_score(y_test_ohe, yp, average='macro', multi_class='ovr')
except:
    auc_val = 0.0
se = recall_score(y_test, ypc, labels=acute_idx, average='macro', zero_division=0)

print(f"  Accuracy:    {ta:.4f}")
print(f"  Macro F1:    {f1:.4f}")
print(f"  AUC:         {auc_val:.4f}")
print(f"  Top-3:       {t3:.4f}")
print(f"  Top-5:       {t5:.4f}")
print(f"  Acute Sens:  {se:.4f}")
print(f"  Temperature: {opt_temp:.2f}")

print(f"\n  CLASSIFICATION REPORT:")
print(classification_report(y_test, ypc, target_names=le.classes_))


# ═══════ 14. SAVE EVERYTHING ═══════
print(f"\n{'='*60}\n  Saving model + artifacts\n{'='*60}")
model.save(f'{SAVE_DIR}/final_symptom_model.keras')
with open(f'{SAVE_DIR}/symptom_scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
with open(f'{SAVE_DIR}/symptom_label_encoder.pkl', 'wb') as f:
    pickle.dump(le, f)
with open(f'{SAVE_DIR}/symptom_temperature.pkl', 'wb') as f:
    pickle.dump(opt_temp, f)
print(f"  ✅ final_symptom_model.keras")
print(f"  ✅ symptom_scaler.pkl")
print(f"  ✅ symptom_label_encoder.pkl")
print(f"  ✅ symptom_temperature.pkl")


# ═══════ 15. PLOTS ═══════
# Training curves
fig, (a1, a2) = plt.subplots(1, 2, figsize=(14, 5))
a1.plot(history.history['loss'], label='Train')
a1.plot(history.history['val_loss'], label='Val')
a1.plot(test_cb.tl, label='Test')
a1.set_title('Loss')
a1.legend()
a1.grid(alpha=0.3)
a2.plot(history.history['accuracy'], label='Train')
a2.plot(history.history['val_accuracy'], label='Val')
a2.plot(test_cb.ta, label='Test')
a2.set_title('Accuracy')
a2.legend()
a2.grid(alpha=0.3)
plt.suptitle('Training Curves', fontweight='bold')
plt.tight_layout()
plt.savefig(f'{FIG_DIR}/training_curves.png', dpi=200)
plt.show()

# Confusion matrix
cm = confusion_matrix(y_test, ypc)
plt.figure(figsize=(14, 12))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Test Set Confusion Matrix')
plt.xticks(rotation=90, fontsize=7)
plt.yticks(fontsize=7)
plt.tight_layout()
plt.savefig(f'{FIG_DIR}/confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# ROC
ato = to_categorical(y_test, NC)
fpr, tpr = {}, {}
for i in range(NC):
    fpr[i], tpr[i], _ = roc_curve(ato[:, i], yp[:, i])
afpr = np.unique(np.concatenate([fpr[i] for i in range(NC)]))
mtpr = np.zeros_like(afpr)
for i in range(NC):
    mtpr += np.interp(afpr, fpr[i], tpr[i])
mtpr /= NC
mauc = sk_auc(afpr, mtpr)
plt.figure(figsize=(8, 7))
plt.plot(afpr, mtpr, color='#2E75B6', lw=3, label=f'Macro ROC (AUC={mauc:.4f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve', fontweight='bold')
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f'{FIG_DIR}/roc_curve.png', dpi=300, bbox_inches='tight')
plt.show()

# LR schedule
plt.figure(figsize=(8, 4))
plt.plot(cosine_lr.lr_history, 'b-', lw=2)
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Cosine LR Schedule')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f'{FIG_DIR}/lr_schedule.png', dpi=200)
plt.show()

# Save metrics
metrics = {
    'test_acc': ta, 'test_loss': tl, 'macro_f1': f1,
    'roc_auc': auc_val, 'top3': t3, 'top5': t5,
    'acute_sens': se, 'temperature': opt_temp,
    'epochs_trained': len(history.history['loss']),
    'val_acc_best': max(history.history['val_accuracy']),
}
pd.DataFrame([metrics]).to_csv(f'{RES_DIR}/final_metrics.csv', index=False, float_format='%.6f')

print(f"\n{'='*60}")
print(f"  ✅ TRAINING COMPLETE")
print(f"{'='*60}")
print(f"  Models  → {SAVE_DIR}/")
print(f"  Figures → {FIG_DIR}/")
print(f"  Results → {RES_DIR}/")
print(f"\n  Download these 4 files for stress testing:")
print(f"    1. {SAVE_DIR}/final_symptom_model.keras")
print(f"    2. {SAVE_DIR}/symptom_scaler.pkl")
print(f"    3. {SAVE_DIR}/symptom_label_encoder.pkl")
print(f"    4. {SAVE_DIR}/symptom_temperature.pkl")
```
